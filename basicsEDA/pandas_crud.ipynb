```python
import pandas as pd
```

```python
# Read
df = pd.read_csv(r"../databases/ecommerce_sales_data.csv") #Read the csv file
df.head(2) # Prints the first two rows
```

```python
# Create
new_row = {'OrderID': "ORD0017", 'CustomerID': "CUST117"}
# Create a new row as a dictionary -> keys are column names, values are the data.

df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
# Append the new row to the existing DataFrame using pd.concat().
# 'ignore_index=True' resets the row index after adding, so the index stays continuous (0, 1, 2, ...).

df.tail() # Prints the last five rows
```

```python
# Read
print(df.iloc[7]) # access by integer position
print(df.loc[:,"Age"]) # access by label or index here, all rows Age cols
print(df.head(1)) # access the first row
print(df.tail(1)) # access the last row 
```

```python
#Update by a specific value
df.loc[df['OrderID']=="ORD00001","CustomerName"]="Jamiya Begam K"
df.iloc[0] V02111v0211
```

```python
#Delete by row
df = df.drop(index=1)
df.head(2)
```

```python
#Delete by column
df = df.drop('State',axis = 1) # here the axis define the row(0) and col(1)
df.head(0) # Prints the cols for validation
```

```python
#Rename cols
df.rename(columns={"City":"MyCity","UnitPrice":"PricePerQuantity"},inplace = True) # Rename the df without creating a copy by using inplace=True
df.head(0) # Prints the cols for validation 
```

```python
# Convert 'OrderDate' column from object/string to datetime 
df["OrderDate"] = pd.to_datetime(df["OrderDate"], format="%Y-%m-%d", errors="coerce")
# Step 1: 'pd.to_datetime()' converts strings or objects into datetime format
# Step 2: 'format' specifies the expected date format (optional but faster and safer)
# Step 3: 'errors="coerce"' will replace invalid dates with NaT (Not a Time) instead of crashing
df.info() # Can be able to see the Dtype here
```

```python
#Rename cols
df.rename(columns={'Sales':'TotalSales','OrderDate':'Date'}, inplace=True)
df.head(0)
```
