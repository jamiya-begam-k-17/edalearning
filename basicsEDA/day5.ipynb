```python
df['Age'].var()   # how spread out the numbers are from their average (mean).
# If the variance is small, it means the ages are all close to each other.
# If the variance is large, it means the ages vary a lot.
```

```python
df['Age'].std()   # Square root of variance.
# Small std → Ages are close to each other (less variation)
# Large std → Ages are very different from each other (more variation)
```

```python
# Interquartile Range (IQR).
IQR = df['Age'].quantile(0.75) - df['Age'].quantile(0.25)   # IQR = Q3 - Q1
# If IQR is small → Ages are tightly packed in the middle
# If IQR is big → There’s a wide spread in the central ages
IQR
```

```python
df[['Age','FinalPrice']].cov()   # Covariance tells you the direction of relationship, NOT the strength.
```

```python
df[['Age','FinalPrice']].corr()   # corr() tells you how strongly Age and FinalPrice are connected and in what direction.
```

```python
# Skew = which side the tail is;
# Kurtosis = how big the tail is.
```

```python
df.skew(numeric_only=True)   # Skewness tells you whether the data is symmetric or leaning to one side.
```

```python
df['Age'].skew()
```

```python
df.kurtosis(numeric_only=True)   # how heavy or light the tails of the distribution are compared to a normal bell curve.
```
