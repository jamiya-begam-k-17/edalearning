```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # for data visualization
import seaborn as sns
```

```python
df = pd.read_csv(r"../databases/ecommerce_sales_data.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
#Change the type of variable (from object to datetime)
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
```

```python
df.info()
```

```python
df_sorted = df.sort_values("OrderDate")
df_sorted
```

```python
plt.figure(figsize=(10,6))
plt.plot(df_sorted["OrderDate"], df["TotalPrice"], marker = "v", linestyle="--", color = 'r', label="Total Price",markersize=6, linewidth=1.5, alpha=0.9)
plt.xlabel("Order Date")
plt.ylabel("Total Price")
plt.legend()
plt.yticks(rotation = 45)
# plt.tight_layout()
plt.show()
```

```python
#Bar chart
df_sorted = df.sort_values("OrderDate")
df_top10 = df_sorted.head(10)

plt.figure(figsize=(10,4))
plt.bar(df_top10["OrderDate"], df_top10["TotalPrice"], color='#9457c2', label="Total Price")
plt.xlabel("Total Price")
plt.ylabel("Order date")
plt.title("Sales over time top 10")
plt.legend()
plt.grid(False)
plt.xticks(rotation = 45)
plt.yticks(rotation = 30)
plt.show()
```

```python
#Scatter chart
plt.figure(figsize=(8,6))
plt.scatter(df["CustomerID"], df["Age"], alpha=0.6, edgecolors='w', s = 60)
# plt.scatter(df["CustomerID"], df["Age"], alpha=0.9, c = "yellow", edgecolors='w', s = 60)
plt.xlabel("Quantity Sold")
plt.ylabel("Sales Amount")
plt.title("Sales vs Quantity")
plt.grid(True)
plt.show()
```

```python
#Histogram
plt.figure(figsize=(10,6))
plt.hist(df["DeliveryDays"], bins=20, color="orange", edgecolor="black", alpha=0.9)
plt.title("Delivery days - Histogram")
plt.xlabel("Delivery Days")
plt.xticks(rotation=50)
plt.show()
```

```python
#Boxplot
plt.figure(figsize=(10,6))
plt.boxplot([df[df["Category"]=="Food"]["TotalPrice"],
            df[df["Category"]=="Toys"]["TotalPrice"],
            df[df["Category"]=="Beauty"]["TotalPrice"]],
            labels = ["Food", "Beauty", "Toys"],
            patch_artist=True, boxprops=dict(facecolor="lightblue", color="blue"),
            medianprops=dict(color="red"))
plt.title("Box plot")
plt.ylabel("Total Price")
plt.show()
```

```python
pycount = df["PaymentMethod"].value_counts()
plt.figure(figsize=(8,4))
plt.pie(pycount,labels=pycount.index,autopct="%1.1f%%", startangle=180)
plt.title("Piechart")
plt.show()
```

```python
# This code creates an **area chart** showing how the total quantity sold for each product category changes month by month.
cat = df.groupby([df['OrderDate'].str[:7], "Category"])["Quantity"].sum().unstack()
cat.plot.area(figsize=(10,6), colormap="Spectral",alpha=0.8) #color = rainbow, viridis, plasma, tab10 and Set2
plt.title("Area plot")
plt.xlabel("month")
plt.ylabel("quanity")
plt.show()
```

```python
corr = df.select_dtypes(include='number').corr() #Select the numeric cols from df
plt.figure(figsize=(10,6)) #
sns.heatmap(corr,annot=True, fmt=".3f", cmap="coolwarm", cbar="True")
plt.title("Correlation map")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=df, x="State", y="City", hue="Category", split=True, inner="quartile", palette="muted", alpha=0.5)
plt.title("Violin Plot")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(data=df, x="Category", y="FinalPrice", hue="PaymentMethod", dodge=True, palette="dark")
plt.title("Swarm Plot")
plt.show()
```

```python
sns.pairplot(df.select_dtypes(include="number").dropna(), diag_kind="kde",markers="*", plot_kws={"alpha":0.08,"s":25})
plt.suptitle("pairplot ", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10,6))
months = df["OrderDate"].str[:7].value_counts().sort_index()
categ = df["Category"].unique()
data = [df[df["Category"] == cat]["OrderDate"].str[:7].value_counts().reindex(months.index, fill_value=0) for cat in categ]
plt.stackplot(months.index,data,labels=categ, alpha = 0.7)
plt.xtickes(rotation=46)
plt.title("Stackplot of Monthly Orders by Category")
plt.legend(loc='upper-left')
plt.show()
```
