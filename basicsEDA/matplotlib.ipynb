```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # for data visualization
```

```python
df = pd.read_csv(r"../databases/ecommerce_sales_data.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
#Change the type of variable (from object to datetime)
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
```

```python
df.info()
```

```python
df_sorted = df.sort_values("OrderDate")
df_sorted
```

```python
plt.figure(figsize=(10,6))
plt.plot(df_sorted["OrderDate"], df["TotalPrice"], marker = "v", linestyle="--", color = 'r', label="Total Price",markersize=6, linewidth=1.5, alpha=0.9)
plt.xlabel("Order Date")
plt.ylabel("Total Price")
plt.legend()
plt.yticks(rotation = 45)
# plt.tight_layout()
plt.show()
```

```python
#Bar chart
df_sorted = df.sort_values("OrderDate")
df_top10 = df_sorted.head(10)

plt.figure(figsize=(10,4))
plt.bar(df_top10["OrderDate"], df_top10["TotalPrice"], color='#9457c2', label="Total Price")
plt.xlabel("Total Price")
plt.ylabel("Order date")
plt.title("Sales over time top 10")
plt.legend()
plt.grid(False)
plt.xticks(rotation = 45)
plt.yticks(rotation = 30)
plt.show()
```

```python
#Scatter chart
plt.figure(figsize=(8,6))
plt.scatter(df["CustomerID"], df["Age"], alpha=0.6, edgecolors='w', s = 60)
# plt.scatter(df["CustomerID"], df["Age"], alpha=0.9, c = "yellow", edgecolors='w', s = 60)
plt.xlabel("Quantity Sold")
plt.ylabel("Sales Amount")
plt.title("Sales vs Quantity")
plt.grid(True)
plt.show()
```

```python
#Histogram
plt.figure(figsize=(10,6))
plt.hist(df["DeliveryDays"], bins=20, color="orange", edgecolor="black", alpha=0.9)
plt.title("Delivery days - Histogram")
plt.xlabel("Delivery Days")
plt.xticks(rotation=50)
plt.show()
```

```python
#Boxplot
plt.figure(figsize=(10,6))
plt.boxplot([df[df["Category"]=="Food"]["TotalPrice"],
            df[df["Category"]=="Toys"]["TotalPrice"],
            df[df["Category"]=="Beauty"]["TotalPrice"]],
            labels = ["Food", "Beauty", "Toys"],
            patch_artist=True, boxprops=dict(facecolor="lightblue", color="blue"),
            medianprops=dict(color="red"))
plt.title("Box plot")
plt.ylabel("Total Price")
plt.show()
```
