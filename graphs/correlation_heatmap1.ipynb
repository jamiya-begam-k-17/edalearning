```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../databases/DA314_S5_GlamNest_Data_Practice.csv")
```

```python
fa = df.loc[df['status']=='Completed'][['price','discount%','discounted_amount(INR)','final_amount','service_rating']]
cor = fa.corr()
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(cor,annot=True,fmt='.3f',cmap='coolwarm',cbar=True)
plt.title("Correlation Heatmap")
plt.show()
```
