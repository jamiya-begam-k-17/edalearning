```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # for data visualization
import seaborn as sns
```

```python
df = pd.read_csv("../databases/DA314_S4_OrderDetails_Data_Concept.csv")
```

```python
ordr_amt=df[df["order_status"]!="Cancelled"]["total_amount"]
```

```python
plt.figure(figsize=(6,4))
sns.histplot(ordr_amt,bins=10,color="skyblue",edgecolor="black",kde=True,alpha=0.5) #Kernel Density Estimation
plt.title("Order Amount Distribution")
plt.xlabel("total_amount")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=ordr_amt,color="lightcoral",fliersize=3,linewidth=2)
plt.title("Order Amount Spread")
plt.xlabel("Order Amount (â‚¹)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(ordr_amt,bins=10,color="skyblue",edgecolor="black",kde="True",alpha=0.5)
plt.title("Order Amount Distribution",weight="bold",fontsize=14)
plt.xlabel("total_amount",weight="bold")
plt.grid()
plt.ylabel("count",weight="bold")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=ordr_amt,color="lightcoral",fliersize=3,linewidth=2)
plt.title("Order Amount Spread",weight="bold",fontsize=14)
plt.grid()
plt.xlabel("Order Amount (â‚¹)")
plt.show()
```

```python
import numpy as np

Q1 = np.percentile(ordr_amt, 25)
Q3 = np.percentile(ordr_amt, 75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = ordr_amt[(ordr_amt < lower) | (ordr_amt > upper)]
print("Number of outliers:", len(outliers))
print(outliers)
```
