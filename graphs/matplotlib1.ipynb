```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../databases/DA314_S4_OrderDetails_Data_Concept.csv")
```

```python
pie_c = df['payment_method'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(
    pie_c,
    labels=pie_c.index,
    startangle=140,
    autopct="%1.1f%%",
    colors=["lightgreen", "lightpink", "lightgrey", "lightseagreen"]
)
plt.title("Payment Method Share")
plt.show()
```

```python
city=df.groupby("city_restaurant")['total_amount'].sum()
city.sort_values(ascending=True).plot(kind='barh',color='skyblue',width=0.8)
plt.xlabel('City')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.title('Total Amount by City')
plt.show()
```

```python
city=df.groupby("city_restaurant")['total_amount'].sum()
city.sort_values(ascending=True).plot(kind='barh',color='darkblue',width=0.8)
plt.xlabel('City')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.title('Total Amount by City')
plt.show()
```

```python
df['order_date']=pd.to_datetime(df['order_date'],errors='coerce')
df['order_month']=df['order_date'].dt.to_period('M')
month_order=df.groupby('order_month').size().sort_index()
plt.figure(figsize=(9,6))
plt.plot(month_order.index.astype(str),month_order.values,marker='o')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.title('Monthly Trend of Order Count')
plt.xticks(rotation=45)
# plt.grid(True,axis='y')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['rating'],y=df['total_amount'],color='green',alpha=0.7)
plt.xlabel('Rating')
plt.ylabel('Total_Amount')
plt.title('Rating vs Total Amount')
plt.tight_layout()
plt.show()
```

```python
city_pay = df.groupby(['city_user','payment_method'])['total_amount'].sum().unstack()
amount = df.groupby('city_user')['total_amount'].sum().sort_values()
city_pay = city_pay.loc[amount.index]
city_pay.plot(kind='bar', stacked=True, figsize=(9,6))
plt.ylabel("Total Amount ()", fontsize=11)
plt.xlabel("City", fontsize=11)
plt.title("City-wise Total Amount by Payment Method", fontsize=13)
plt.legend(title="Payment Method")
plt.show()
```
