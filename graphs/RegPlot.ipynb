```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../databases/DA314_S4_OrderDetails_Data_Concept.csv")
```

```python
ordr_amt=df[df["order_status"]!="Cancelled"]
```

```python
corr = ordr_amt[['avg_cost_for_two','rating']].corr()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(corr,annot=True,fmt=".3f",cmap="coolwarm",cbar=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(
    data=df,
    x='avg_cost_for_two',
    y='rating',
    scatter_kws={'alpha':0.6, 'color':'orange'},
    line_kws={'color':'red'},
)
plt.title("Cost vs. Customer Rating with Trend Line", fontsize=14,weight="bold")
plt.xlabel("Average Cost for Two ()")
plt.ylabel("Customer Rating(1-5)")
plt.grid()
plt.tight_layout()
plt.show()
```
