```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../databases/DA314_S5_GlamNest_Data_Practice.csv")
```

```python
fa = df.loc[df['status']=='Completed']['final_amount']
```

```python
plt.figure(figsize=(9,4))
sns.histplot(fa,bins=11,color="skyblue",edgecolor="black",kde=True,alpha=0.6)
plt.title("Service Amount Distribution")
plt.xlabel("final_amount")
plt.ylabel("Count")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x=fa,color="lightcoral",fliersize=3,linewidth=2)
plt.title("Service Amount Spread")
plt.xlabel('final_amount')
plt.grid()
plt.show()
```

```python
service_rating = df.loc[df['status']=='Completed', 'service_rating']
```

```python
plt.figure(figsize=(9,4))
sns.histplot(service_rating,bins=13,color="salmon",edgecolor="black",kde=True,alpha=0.6)
plt.title("Service Rating Distribution")
plt.xlabel("service_rating")
plt.ylabel("Count")
# plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(x=service_rating,color='peachpuff',fliersize=3,linewidth=2)
plt.title("Service Rating Spread")
plt.xlabel("service_rating")
plt.grid()
plt.show()
```
